```python
import nltk
from nltk.corpus import wordnet
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
import pandas as pd
from bs4 import BeautifulSoup
import requests
from textblob import TextBlob
import re
import traceback

# Function to extract article text from URL
def extract_article_text(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract the article text
        article_text = ""
        if soup.find("article"):
            article = soup.find("article")
            if article.find("h1"):
                article_text += article.find("h1").get_text() + "\n"
            for paragraph in article.find_all("p"):
                article_text += paragraph.get_text() + "\n"

        return article_text
    except requests.exceptions.RequestException as e:
        print(f"Error extracting article text from URL: {url}")
        print(f"Error message: {str(e)}")
        traceback.print_exc()
        return None
    except Exception as e:
        print(f"An error occurred while extracting article text from URL: {url}")
        print(f"Error message: {str(e)}")
        traceback.print_exc()
        return None


# Load the input data from the "Input.xlsx" file
input_file = "Input.xlsx"
input_data = pd.read_excel(input_file)

# Initialize the output data DataFrame with appropriate columns
output_data = pd.DataFrame(columns=["URL", "Title", "POSITIVE SCORE", "NEGATIVE SCORE", "POLARITY SCORE", "SUBJECTIVITY SCORE", "AVG SENTENCE LENGTH", "PERCENTAGE OF COMPLEX WORDS", "FOG INDEX", "AVG NUMBER OF WORDS PER SENTENCE", "COMPLEX WORD COUNT", "WORD COUNT", "SYLLABLE PER WORD", "PERSONAL PRONOUNS", "AVG WORD LENGTH"])

# Load the Stop Words List
stop_words_file = "StopWords/StopWords_Generic.txt"
with open(stop_words_file, "r") as file:
    stop_words = file.read().splitlines()

# Load the positive and negative words lists
positive_words_file = "MasterDictionary/positive-words.txt"
with open(positive_words_file, "r") as file:
    positive_words = file.read().splitlines()

negative_words_file = "MasterDictionary/negative-words.txt"
with open(negative_words_file, "r") as file:
    negative_words = file.read().splitlines()

# Function to clean the text and calculate text analysis variables
def analyze_text(text):
    # Clean the text by removing stop words
    cleaned_text = " ".join(word for word in text.split() if word.lower() not in stop_words)

    # Sentiment Analysis
    blob = TextBlob(cleaned_text)
    sentiment_score = blob.sentiment.polarity

    # Count positive and negative words
    positive_count = sum(1 for word in cleaned_text.split() if word.lower() in positive_words)
    negative_count = sum(1 for word in cleaned_text.split() if word.lower() in negative_words)

    # Rest of your analysis code...

    return sentiment_score, positive_count, negative_count

# Iterate over each row in the input data
for index, row in input_data.iterrows():
    url = row["URL"]
    print(f"Processing URL: {url}")
    article_text = extract_article_text(url)
    
    # Perform text analysis if article text is not None
    if article_text is not None:
        sentiment_score, positive_count, negative_count = analyze_text(article_text)
        print("Text analysis completed.")
        
        # Create a new row with the analysis results
        new_row = [url, None, sentiment_score, positive_count, negative_count] + [None] * (len(output_data.columns) - 5)
        
        # Append the new row to the output data
        output_data.loc[len(output_data)] = new_row
    else:
        print("Skipping analysis for URL due to extraction error.")

# Save the output data to the "Output Data Structure.xlsx" file
output_file = "Output Data Structure.xlsx"
output_data.to_excel(output_file, index=False)

print("Analysis completed. Output saved to Output Data Structure.xlsx.")
```
