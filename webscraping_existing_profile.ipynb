```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
#chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')

# First Profile
user_data_directory_1 = r'C:\Users\abhi2\OneDrive\Documents\portfolio_projects\profiles'
profile_name_1 = r'Profile 3'
chrome_options.add_argument(f'--user-data-dir={user_data_directory_1}')
chrome_options.add_argument(f'--profile-directory={profile_name_1}')

# Set up Chrome service for the first profile
service1 = Service(ChromeDriverManager().install())

# Create Chrome driver instance for the first profile
driver1 = webdriver.Chrome(service=service1, options=chrome_options)

# Navigate to Google and search for the query
query = 'site:youtube.com openinapp.co'
url = f"https://www.google.com/search?q={query}&num=10000"

driver1.get(url)
time.sleep(10)

# Extract the links from the search results
search_results = driver1.find_elements(By.CSS_SELECTOR, ".yuRUbf > a")

print(len(search_results))  # Print the length of search_results

links = []
for result in search_results:
    link = result.get_attribute('href')
    if 'youtube.com/' in link:
        links.append(link)
        print(link)
    if len(links) >= 10000:
        break

# Create a Pandas dataframe with the links
df = pd.DataFrame({'links': links})
print(df.head())
df.to_csv('youtube_links_1.csv', index=False)

# Delay before quitting the first driver
time.sleep(5)

# Second Profile
user_data_directory_2 = r'C:\Users\abhi2\OneDrive\Documents\portfolio_projects\profiles'
profile_name_2 = r'Profile 5_Tan'

# Create a new Chrome options object for the second profile
chrome_options2 = Options()
chrome_options2.add_argument('--no-sandbox')
chrome_options2.add_argument('--disable-dev-shm-usage')
#chrome_options2.add_argument('--headless')
chrome_options2.add_argument('--disable-gpu')
chrome_options2.add_argument(f'--user-data-dir={user_data_directory_2}')
chrome_options2.add_argument(f'--profile-directory={profile_name_2}')

# Set up Chrome service for the second profile
service2 = Service(ChromeDriverManager().install())

# Create Chrome driver instance for the second profile
driver2 = webdriver.Chrome(service=service2, options=chrome_options2)

# Navigate to Google and search for the query (second profile)
driver2.get(url)
time.sleep(10)

# Extract the links from the search results (second profile)
search_results = driver2.find_elements(By.CSS_SELECTOR, ".yuRUbf > a")

print(len(search_results))  # Print the length of search_results

links = []
for result in search_results:
    link = result.get_attribute('href')
    if 'youtube.com/' in link:
        links.append(link)
        print(link)
    if len(links) >= 10000:
        break

# Create a Pandas dataframe with the links (second profile)
df = pd.DataFrame({'links': links})
print(df.head())
df.to_csv('youtube_links_2.csv', index=False)

# Delay before quitting the second driver
time.sleep(5)

# Quit the drivers
driver1.quit()
driver2.quit()
```
