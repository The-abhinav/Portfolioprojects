```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import undetected_chromedriver as uc
import concurrent.futures

# Set up Chrome options for Profile 1
chrome_options1 = uc.ChromeOptions()
chrome_options1.add_argument('--no-sandbox')
chrome_options1.add_argument('--disable-dev-shm-usage')

# Set up Chrome options for Profile 2
chrome_options2 = uc.ChromeOptions()
chrome_options2.add_argument('--no-sandbox')
chrome_options2.add_argument('--disable-dev-shm-usage')

# Function to scrape links from Google search
def scrape_links(query, chrome_options):
    service = uc.ChromeService(ChromeDriverManager().install())
    driver = uc.Chrome(service=service, options=chrome_options)
    query_url = f"https://www.google.com/search?q={query}&num=10000"
    driver.get(query_url)
    time.sleep(10)
    search_results = driver.find_elements(By.CSS_SELECTOR, ".yuRUbf > a")
    links = []
    for result in search_results:
        link = result.get_attribute('href')
        if 'youtube.com/' in link:
            links.append(link)
            print(link)
        if len(links) >= 10000:
            break
    driver.quit()
    return links

# Profile 1 - Visit Google and perform search
query1 = 'site:youtube.com openinapp.co'

# Profile 2 - Visit Google and search for pasta
query2 = 'site:youtube.com pasta'

# Create a ThreadPoolExecutor with max_workers=2
executor = concurrent.futures.ThreadPoolExecutor(max_workers=2)

# Submit tasks to run the scraping tasks
task1 = executor.submit(scrape_links, query1, chrome_options1)
task2 = executor.submit(scrape_links, query2, chrome_options2)

# Wait for tasks to complete
concurrent.futures.wait([task1, task2])

# Retrieve the results from the tasks
links1 = task1.result()
links2 = task2.result()

# Create Pandas dataframes with the links from Profile 1 and Profile 2
df1 = pd.DataFrame({'links': links1})
df2 = pd.DataFrame({'links': links2})

# Print the number of links scraped for each profile
print(f"Number of links scraped for Profile 1: {len(df1)}")
print(f"Number of links scraped for Profile 2: {len(df2)}")
```
